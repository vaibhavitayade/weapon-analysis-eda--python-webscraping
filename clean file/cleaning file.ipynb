```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import re

def url():
    url = "https://www.airsoftstation.com"  
    weapons = []
    
    try:
        print(f" Trying {url}...")
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}, timeout=10)
        
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            products = soup.find_all(['div', 'li'], class_=lambda x: x and 'product' in str(x).lower())
            
            for product in products[:400]:  
                try:
                    name = product.find(['h2', 'h3', 'h4']).get_text(strip=True) if product.find(['h2', 'h3', 'h4']) else None
                    price = product.find(class_=lambda x: x and 'price' in str(x).lower()).get_text(strip=True) if product.find(class_=lambda x: x and 'price' in str(x).lower()) else None
                    
                    if name and price:
                        weapons.append({
                            'Name': name,
                            'Type': random.choice(['Rifle', 'Pistol', 'Shotgun', 'SMG', 'Sniper']),
                            'Price': price,
                            'Weight': f'{random.uniform(1.5, 8.0):.1f}kg',
                            'Size': f'{random.randint(40, 120)}cm',
                            'Caliber': random.choice(['5.56mm', '9mm', '7.62mm', '.45 ACP', '12 Gauge']),
                            'Range': f'{random.randint(50, 800)}m',
                            'Availability': random.choice(['In Stock', 'Out of Stock']),
                            'Brand': random.choice(['CYMA', 'G&G', 'Tokyo Marui', 'Lancer Tactical']),
                            'FPS': f"{random.randint(250, 450)} FPS",
                            'Magazine Capacity': f"{random.randint(20, 200)} rounds",
                            'Power Source': random.choice(['Electric', 'Gas', 'CO2', 'Spring']),
                            'Customer Rating': round(random.uniform(1.0, 5.0), 1)
                        })
                except:
                    continue
    except:
        print(" Failed to scrape site")
    
    return weapons

def generate_fallback_data():
    weapons = []
    types = ['Rifle', 'Pistol', 'Shotgun', 'SMG', 'Sniper']
    calibers = ['5.56mm', '9mm', '7.62mm', '.45 ACP', '12 Gauge']
    
    for i in range(400):
        weapon_type = random.choice(types)
        weapons.append({
            'Name': f'{weapon_type} M{random.randint(100,999)}',
            'Type': random.choice(['Rifle', 'Pistol', 'Shotgun', 'SMG', 'Sniper']),
            'Price': f'${random.randint(300, 2500)}',
            'Length (cm)': f'{random.randint(40, 120)}cm',
            'Caliber': random.choice(calibers),
            'Range': f'{random.randint(50, 800)}m',
            'Availability': random.choice(['In Stock', 'Out of Stock']),
            'Brand': random.choice(['CYMA', 'G&G', 'Tokyo Marui', 'Lancer Tactical']),
            'FPS': f"{random.randint(250, 450)} FPS",
            'Magazine Capacity': f"{random.randint(20, 200)} rounds",
            'Power Source': random.choice(['Electric', 'Gas', 'CO2', 'Spring']),
            'Customer Rating': round(random.uniform(1.0, 5.0), 1)
        })
    return weapons


print("Starting web scraping...")
data = url()

if len(data) < 400:
    print(f" Only got {len(data)} products, generating sample data instead...")
    data = generate_fallback_data()

df = pd.DataFrame(data[:400])

# ---------- Clean Data ----------

# 1. Convert Price → Cost (₹) 
usd_to_inr = 83  # approx conversion rate
df['Cost'] = (
    df['Price']
    .str.replace(r'[^0-9.]', '', regex=True)   # remove $ or text
    .astype(float) * usd_to_inr
).round(2)
df = df.drop(columns=['Price'])

# 2. Clean Weight (kg)
if 'Weight' in df.columns:
    df['Weight'] = df['Weight'].str.replace(r'[^0-9.]', '', regex=True).astype(float)

# 4. Clean Range (m)
if 'Range' in df.columns:
    df['Range'] = df['Range'].str.replace(r'[^0-9.]', '', regex=True).astype(int)

# 5. Standardize Caliber
df['Caliber'] = (
    df['Caliber']
      .astype(str)
      .str.lower()
      .replace({
          ".45 acp": "11.43mm",
          ".380 acp": "9mm",
          ".32 acp": "7.65mm",
          ".25 acp": "6.35mm",
          "12 gauge": "18.53gauge",
          "20 gauge": "15.63gauge"
      })
)

# Split into Diameter + Type
df[['Caliber_Diameter', 'Caliber_Type']] = df['Caliber'].str.extract(r'(\d+\.?\d*)([a-zA-Z]+)')
df['Caliber_Diameter'] = df['Caliber_Diameter'].astype(float)

# Drop old Caliber
df = df.drop(columns=['Caliber'])

# ---------- Save File ----------
df.to_csv("weaponsdata_clean2.csv", index=False)

print(f"Saved {len(df)} rows to weaponsdata_clean.csv")
print("Columns:", list(df.columns))
```
